```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the csv file
turbidity_bouy = pd.read_csv('Turbidity_Buoy.csv',skiprows=1)  

# convert the timestamp to datetime in pandas
turbidity_bouy['Timestamp (UTC+12:00)'] = pd.to_datetime(turbidity_bouy['Timestamp (UTC+12:00)'])

# Set the 'Timestamp (UTC+12:00)' column as the index
turbidity_bouy.set_index('Timestamp (UTC+12:00)', inplace=True)

```

```python
print(turbidity_bouy.describe())
```

```python
# Read the csv file
turbidity_platform = pd.read_csv('Turbidity_Platform.csv',skiprows=1)  

# convert the timestamp to datetime in pandas
turbidity_platform['Timestamp (UTC+12:00)'] = pd.to_datetime(turbidity_platform['Timestamp (UTC+12:00)'])

# Set the 'Timestamp (UTC+12:00)' column as the index
turbidity_platform.set_index('Timestamp (UTC+12:00)', inplace=True)
```

```python
turbidity_platform.describe()
```

```python
# Plotting time series data
plt.figure(figsize=(10,6))
plt.plot(np.log(turbidity_bouy['Value (NTU)']), linestyle='-')
plt.xlabel('Timestamp (UTC+12:00)')
plt.ylabel('Value (NTU)(log)')
plt.title('Turbidity Buoy')
plt.grid(True)
plt.show()
```
